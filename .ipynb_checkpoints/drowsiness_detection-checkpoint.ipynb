```python
import numpy as np 
import pandas as pd 
import os
import tensorflow as tf

from glob import glob
from tqdm import tqdm

import matplotlib.pyplot as plt
from PIL import Image
```

## Loading the images 

```python
X = []
Y = []

for i in tqdm(glob('C:/Users/gowth/Documents/Drowsiness-Detection/train/Open_Eyes/*')):
    temp = np.array(Image.open(i).resize((64,64)))
    X.append(temp)
    Y.append(1)
    
for i in tqdm(glob('C:/Users/gowth/Documents/Drowsiness-Detection/train/Closed_Eyes/*')):
    temp = np.array(Image.open(i).resize((64, 64)))
    X.append(temp)
    Y.append(0)
    
```

## Normalizing the images(with any algorithm)

```python
# X = np.array(X)
# X = X/255.0
# Y = np.array(Y)
```

## Normalizing the images using Min-Max Scaling algorithm

```python
X = (np.array(X) - np.min(X)) / (np.max(X) - np.min(X))
X = X/255.0
Y = (np.array(Y) - np.min(Y)) / (np.max(Y) - np.min(Y))
```

## Expanding the dimensions of array(numpy arr)


```python
X = np.expand_dims(X, -1)
```

## Spliting the dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

##  CNN Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Conv2D, BatchNormalization, MaxPooling2D, Dropout, Flatten

model = tf.keras.models.Sequential([Input(shape=(64, 64, 1)),
                        
        Conv2D(filters = 32, kernel_size = 5, strides = 1, activation = 'relu'),
        Conv2D(filters = 32, kernel_size = 5, strides = 1, activation = 'relu', use_bias=False),
        BatchNormalization(),
        MaxPooling2D(strides = 2),
        Dropout(0.3),
        Conv2D(filters = 64, kernel_size = 3, strides = 1, activation = 'relu'),
        Conv2D(filters = 64, kernel_size = 3, strides = 1, activation = 'relu', use_bias=False),
        BatchNormalization(),
        MaxPooling2D(strides = 2),
        Dropout(0.3),

        Flatten(),
        Dense(units = 128, activation = 'relu', use_bias=False),
        BatchNormalization(),
                                    
        Dense(units = 84, use_bias=False, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(units = 1, activation = 'sigmoid')
       ])
```

## Compiling the model 

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Training the model

```python
callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='bestModel.h5',
    save_weights_only=False,
    monitor='val_loss',
    mode='min',
    save_best_only=True,
    verbose=1
)
model.fit(x_train, y_train, validation_split=0.2, epochs=40, batch_size=32, callbacks=callback)
```

## Model Evaluation 

```python
model.evaluate(x_test, y_test)
```

## Loading the Model Performance

```python
from keras.models import load_model
best_model = load_model('./bestModel.h5')
best_model.evaluate(x_test, y_test)
```

## Changing the dimensions and plotting the trained images

```python
for i in x_test[0:5]:
    result = best_model.predict(np.expand_dims(i,0))
    plt.imshow(i)
    plt.show()
    
    if result > 0.5:
        print('Open')
    else:
        print('Closed')
```

## Prediction and Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
plt.figure(figsize=(15, 5))
preds = best_model.predict(x_test)
preds = (preds >= 0.5).astype(np.int32)
cm = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(cm, index=['Closed', 'Open'], columns = ['Closed', 'Open'])
plt.subplot(121)
plt.title("Confusion Matrix\n")
sns.heatmap(df_cm, annot=True, fmt="d", cmap='viridis')
plt.ylabel("Predicted")
plt.xlabel("Actual Outcome")
```

## Accuracy

```python
accuracy = best_model.evaluate(x_test, y_test)[1]
print("Accuracy :", accuracy)
```
